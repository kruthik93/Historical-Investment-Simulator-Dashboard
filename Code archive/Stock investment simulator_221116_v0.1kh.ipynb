# Objective: Create a tool to simulate end result of investing based on investments into certain stocks
## The tool should be able to take in an input bi-weekly or monthly amount and simulate what the final amount would have been if it was invested in one stock or multiple stocks

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# first step is to figure our how to download the stock market data
# I remember something like FRED or CRED?

# Figured out that yahoofinance provides this so importing that
import yfinance as yf
from yahoofinancials import YahooFinancials

#Dont know what this library does but we'll figure out
```

```python
# Downloading a sample stock data - lets take tesla for example
tsla_df = yf.download('TSLA', 
                     start = '2021-01-01',
                     end = '2021-12-31',
                     progress=False)

tsla_df.head(20).round()
```

```python
#Trying to get the date as column
sample_df = tsla_df.reset_index()

sample_df.columns
#Looks like reset_index() works to provide this solution
```

```python
#Trying to streamline the downloading process and create a standardized dataframe with these changes

ticker_list = ['MSFT','TSLA']

start_dt = '2018-01-01'
end_dt = '2022-10-31'

stock_df = pd.DataFrame()

for ticker in ticker_list:
    ticker_df = yf.download(ticker,
                     start = start_dt,
                     end = end_dt)
    ticker_df = ticker_df.reset_index()
    ticker_df['Ticker'] = ticker
    if stock_df.empty:
        print("dataframe is empty")
        stock_df = ticker_df
    else:
        stock_df = pd.concat([stock_df,ticker_df])

stock_df.head(20)
```

```python
stock_df.describe()
```

```python
#Now that we have got the dataframes with the stock market data we need, 1st thing to do is to calculate the average stock price for that day
#Avg stock price we will define as the mean of the high and low value
#np.mean([stock_df['High'],stock_df['Low']])

#need to figure out how to calculate mean row by row
stock_df['Avg'] = stock_df[['High','Low']].mean(axis=1)
```

```python
#Now that the mean has been calculated, we need to see how we can access it for a particular day for a particular stock

stock = 'MSFT'
date = '2018-06-15'

stock_avg_cost = stock_df[(stock_df['Ticker']==stock) & (stock_df['Date']==date)]['Avg']
stock_avg_cost
```

```python
#Great, now we need to understand how much amount was invested and the percentage of it which will be put into this stock
#lets say we invested $500 regulary
reg_inv_amt = 500
#and lets say for this tock it is 100%in that stock
stock_perc = 100
#What is the amount which would have been invested on that day is
stock_inv_amt = reg_inv_amt*(float(stock_perc)/100)
#Printing to check
stock_inv_amt
```

```python
#Great, weve got the investment amount, now to check how much stock it would have bought
stock_qty = float(stock_avg_cost)/stock_inv_amt
stock_qty
```

```python
#now we need to store this in a dataframe for each date it has been bought
#First creating a dictionary out of the current data
stock_buy_dict = {'Date':date,
                  'Stock':stock,
                  'Stock_avg_cost':stock_avg_cost,
                  'Stock_inv_amt':stock_inv_amt,
                  'Stock_qty':stock_qty}

stock_qty_df = pd.DataFrame(stock_buy_dict)
stock_qty_df
```

```python
stock_df.Date.dt.day
```

```python
#Thought of another method, we can bring in investment percentages into the main stock dataset after filtering for dates of investment
#Creating the day number marker as a column
stock_df['Day'] = stock_df.Date.dt.day
stock_df.head(20)

#need to filter for days of investment i.e 15th and 30th

inv_days = [15,30]

stock_inv_days_df = stock_df[stock_df['Day'].isin(inv_days)]

stock_inv_days_df
```

```python
#Adding a column to denote investment amount
stock_inv_days_df['Reg_inv_amt'] = reg_inv_amt
stock_inv_days_df.head()
```

```python
#Creating a dictionary of stocks and investment percentages
inv_mix_dict = {'Ticker':['TSLA','MSFT'],
                'Inv_perc':[0.3,0.7]}
inv_mix_df = pd.DataFrame(inv_mix_dict)
inv_mix_df
```

```python
#Mapping this back to the the investment days dataframe
print(stock_inv_days_df.shape)
stock_inv_days_df = stock_inv_days_df.merge(inv_mix_df,on=['Ticker'],how='left')
print(stock_inv_days_df.shape)
```

```python
#Calcluating other columns required
#Amount invested
stock_inv_days_df['Stock_inv_amt'] = stock_inv_days_df['Reg_inv_amt']*stock_inv_days_df['Inv_perc']
stock_inv_days_df.head()
#Quantity bought
stock_inv_days_df['Stock_qty'] = stock_inv_days_df['Avg']/stock_inv_days_df['Stock_inv_amt']
stock_inv_days_df.head()
```

```python
#Now we need the price of the stocks during the latest day
stock_latest_dt_df = stock_df.groupby('Ticker').agg({'Date':'max'}).reset_index().rename(columns={'Date':'Latest_date'})
stock_latest_dt_df
```

```python
#Bringing out the latest price for the stock
stock_latest_df = stock_latest_dt_df.merge(stock_df[['Ticker','Date','Avg']], left_on = ['Ticker','Latest_date'], right_on = ['Ticker','Date'], how = 'left').reset_index().rename(columns={'Avg':'Latest_avg'})
stock_latest_df
```

```python
#Merging this back to the investment df
stock_inv_days_df = stock_inv_days_df.merge(stock_latest_df[['Ticker','Latest_avg']], on = ['Ticker'], how = 'left')
stock_inv_days_df['Latest_value'] = stock_inv_days_df['Stock_qty']*stock_inv_days_df['Latest_avg']
stock_inv_days_df
```

```python
#Final portfolio vlaue is
stock_inv_days_df['Latest_value'].sum()
```
